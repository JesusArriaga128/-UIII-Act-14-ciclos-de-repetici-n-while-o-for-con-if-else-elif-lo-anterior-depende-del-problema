Descripción del Problema:
Diseña un programa interactivo en Python que funcione como un analizador de texto en tiempo real. El programa debe solicitar al usuario que introduzca un carácter (letra, número o símbolo) de forma continua. Al recibir el carácter, el sistema deberá evaluar de forma automática si este corresponde a una vocal (ya sea mayúscula o minúscula) o no.

```python
print("====================================================")
print("     Detector de Vocales (Presiona Espacio para salir)     ")
print("====================================================")

# Ciclo infinito que se ejecutará hasta que se introduzca un espacio
while True:

    # Pedimos al usuario que ingrese un solo carácter
    caracter = input("Introduce un carácter: ")

    # Condición de salida: Si el usuario presiona la barra espaciadora (un espacio en blanco)
    if caracter == " ":
        print("¡Programa terminado! Gracias.")
        break  # Rompe el ciclo while y finaliza el programa

    # Convertimos el carácter a minúscula para evaluar mayúsculas y minúsculas por igual
    caracter_minuscula = caracter.lower()

    # Evaluamos si el carácter es una vocal
    if caracter_minuscula == "a" or caracter_minuscula == "e" or caracter_minuscula == "i" or caracter_minuscula == "o" or caracter_minuscula == "u":
        print("Resultado: VOCAL")
    else:
        print("Resultado: NO VOCAL")
```
